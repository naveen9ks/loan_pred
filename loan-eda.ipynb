Step 4: Statistical Analysis

4.1 Detect Outliers

4.1.1 Applicant Income Outliers Detection

```python
df.boxplot("ApplicantIncome")       # creates box plot
plt.title("Applicant Income Outliers Detection")        # shows title of chart
plt.show()      # shows entire chart
```

    Inference:
        The Applicant Income data has huge amount of outliers in it
        This has to be removed for getting pure data

4.1.1(a) Clean Outliers for Applicant Income

```python
# Remove outliers by Inter Quantile Range (IQR) method
Q1= df['ApplicantIncome'].quantile(0.25)
Q3= df['ApplicantIncome'].quantile(0.75)

IQR= Q3-Q1

lower= Q1-0.25*IQR
upper= Q3-0.25*IQR

df_cleaned= df[(df['ApplicantIncome']>=lower)&(df['ApplicantIncome']<=upper)]
```

```python
plt.boxplot(df_cleaned['ApplicantIncome'])      # creates boxplot
plt.title("Cleaned Applicant Income")       # shows title of chart
plt.grid()      # shows grid lines
plt.show()      # shows entire graph
```

    Inference:
        The Applicant Income data's outlier's has been detected and removed by Inter Quantile Range (IQR) method

4.1.3 Loan Amount Outlier's Detection

```python
df.boxplot("LoanAmount")        # creates boxplot
plt.title("Loan Amount Outlier's Detection")        # shows title of chart
plt.show()      # shows entire chart
```

4.1.3(a) Cleaned Loan Amount

```python
# Remove outliers by Inter Quantile Range (IQR) method
Q1= df['LoanAmount'].quantile(0.25)
Q3= df['LoanAmount'].quantile(0.75)

IQR= Q3-Q1

lower= Q1-0.5*IQR
upper= Q3-0.5*IQR

df_cleaned= df[(df['LoanAmount']>=lower)&(df['LoanAmount']<=upper)]
```

```python
plt.boxplot(df_cleaned['LoanAmount'])       # creates box plot
plt.title("Cleaned Loan Amount")        # shows title of chart
plt.grid()      # shows grid lines in chart
plt.show()      # shows entire graph
```

    Inference:
        The Loan Amount data has been completely removed outliers by Inter Quantile Range (IQR) method
